```python
import sklearn
```

```python
from sklearn.datasets import load_iris
data = load_iris()
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
for_kartinki = pd.DataFrame(data['data'].tolist(), columns = data['feature_names'])
```

```python
for_kartinki['target'] = data["target"]
```

```python
for_kartinki[['sepal length (cm)','sepal width (cm)']].plot.scatter(x='sepal length (cm)',y = 'sepal width (cm)')
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x='sepal length (cm)',y = 'sepal width (cm)',data = for_kartinki, hue = 'target')
```

При условии использования данной визуализации стоит отметить, что классы хорошо разделимы

```python
sns.countplot(x = 'target',data = for_kartinki, hue = 'target');
```

Данный график демонстрирует распределение по классам. В данном случае количество классов одинаково

```python
for_kartinki['sepal length (cm)'].hist(bins = 20);
```

Этот график показывает распределение столбца. Конкретный график демонстрирует наличие двух пиков. 

```python
sns.heatmap(for_kartinki.corr())
```

Во-первых, это просто красиво). Во-вторых, мы видим сильную корреляцию между длиной и шириной "пестика" и "тычинки"
