Задание 1
Изобразите с помощью matplotlib. 

Изобразите точку x + 2y + 3z. 

Найдите угол между векторами x, y и x, z.

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import
```

```python
y = np.array([0,1,2])
z = np.array([2,1,0])
```

```python
x = -2*y -3*z
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity([x],[y]), cosine_similarity([x],[z])
```

Для матрицы 3x3 можно посмотреть на корни характеристического многочлена, посчитанные в numpy.

```python
m1 = np.array([[2, 2],
               [1, 3]])

m2 = np.array([[4, 1, -1],
               [1, 4, -1],
               [-1, -1, 4]])
```

```python
np.linalg.eig(m1)
```
